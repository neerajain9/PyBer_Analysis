```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
plt.xlabel('Month')
plt.ylabel('Fare in US Dollars')
plt.title('Fares by Month')
plt.show
```

```python
%matplotlib inline
```

```python
## Create chart using the "Object-Oriented Interface Method"
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
## Alternate way of representing OOPs Method
# Create the plot with ax.plt()
fig = plt.figure() ### To change figure-level attributes (such as axis labels, a title, 
                   ### or a legend) or save the figure as an image, use this
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
## 3rd way of representing OOPS Method
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
#plt.plot(x_axis, y_axis, label='Boston')
plt.plot(x_axis, y_axis, color="blue" , linewidth=2, marker="*", label='Boston')


# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Method 1 (use 'o' at the end of the statement otherwise it will generate a line chart)
# plt.plot(x_axis, y_axis, 'o')

# OR

#Method 2
plt.scatter(y_axis, x_axis, c="red", marker = "x", s = y_axis, label="Chigaco")
# (s = size => can equate to data element and size of each data element will vary with the value e.g. s = y_axis )

plt.xlabel("Fares ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")

# Invert months to read from Jan on top
plt.gca().invert_yaxis()

# Set the x limit between 0 and 45.
plt.xlim(5, 50)

# Add grid lines
#plt.grid()

# Add the legend.
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Explode the values to split out of the main chart
explode_values = (0.5, 0 , 0, 0.2, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Increase the size of the plot
plt.subplots(figsize=(8, 8))

plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')
plt.show()
```

```python
## PIE CHART USING OOPS
```

```python
# Explode the values to split out of the main chart
explode_values = (0, 0 , 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "pink", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Increase the size of the plot
#fig, ax = plt.subplots(figsize=(8, 8))
fig, ax = plt.subplots(figsize=(8, 8))

ax.pie(y_axis,
       colors=colors,
       explode=explode_values,
       shadow=True,            # add shadow
       startangle=90,          #so that Jan is at the top
       counterclock = False,   # change direction of months from left to right in clockwise direction
       labels=x_axis,
       autopct='%.1f%%')
plt.show()
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Add Error-Bar
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3) # capsize is to 'cap' the error bars. 
                                                    # Notice the cap at the end of each bar for each data point.
```

```python
# Add Error-Bar (USING OOPS)
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Add Error Bars to a Bar Chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add Error Bars to a Bar Chart (USING OOPS)
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Change the Major Ticks
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Change the Major Ticks (USING OOPS)
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
# Or use
# ax.xaxis.set_ticks(np.arange(0, 51, step=5.0))

plt.show()
```

```python
# Change the MINOR Ticks
from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
